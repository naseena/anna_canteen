```python
#function
def my_fun(df,Muncipal_Corporation,No_Of_Gowdowns):
    newdf = df[(df['Muncipal_Corporation']==Muncipal_Corporation) & (df['No_Of_Gowdowns']==No_Of_Gowdowns)]
    final=newdf.to_dict('dict')
    return final
```

```python
my_fun(Corporation,"Corporation1","Gowdown1")
```
